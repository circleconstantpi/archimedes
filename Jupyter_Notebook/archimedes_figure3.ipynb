```python
# Radian to degree
def rad2deg(rad):
    '''Convert radian to degree'''
    deg = (180 / pi) * rad
    return(deg)

# Degree to radian
def deg2rad(deg):
    '''Convert degree to radian'''
    rad = (pi / 180) * deg
    return(rad)
```

```python
# Determine the intersection point of an equation of a line with a vertical straight line. 
var('x0')
alpha = 30
for i in range(0,5):
    n = 6 * 2^i
    rad = deg2rad(alpha)
    y0 = -1
    f = -x0*tan(rad) == y0 ;f
    e = f.solve(x0, solution_dict=True)
    x1 = (e[0][x0])/n
    y1 = tan(rad)
    print("alpha =", alpha)
    #print(e)
    #print("x =", x1)
    #print("x =", x1.n())
    #print("y =", y1)
    print("y =", y1.n())
    alpha /= 2
```

```python
# Plot a semi-circle with axes.
r = 1
p = plot(sqrt(r^2 - x^2),(x,-1, 1), thickness=1.0, color='magenta', xmin=-1.1, xmax=1.1, 
         ymin=-0.1, ymax=1.1, axes_labels=['$x$','$y$'], zorder=0, aspect_ratio=1)
# Add a straight line.
p += line([(-1.0, 0.0), (-1, 0.58)], color='darkgreen', zorder=0)
# Add the angle bisector.
alpha = 30
for i in range(0,5):
    rad = deg2rad(alpha)
    p += plot(-x*tan(rad),(x, -1.0, 0.0), color='blue', zorder=1)
    alpha /= 2
# Add straight line.
p += plot(-x*tan(-(1/96)*pi),(x, -1.0, 0.0), color='darkorange', zorder=1)
# Add labeling.  
label = {'$O$': (0.05, -0.05), '$A$': (-1.0, -0.12), '$B$': (1.0, -0.12), 
         '$C$': (-1.05, 0.60), '$D$': (-1.05, 0.29), '$E$': (-1.05, 0.15),
         '$F$': (-1.05, 0.09), '$G$': (-1.05, 0.03), '$H$': (-1.05, -0.035)} 
for char, posi in label.items():
    p += text(char, posi, fontsize=14, color='black', zorder=2)
# Add points.
p += point(((1,0), (0,0), (-1,0), (-1, 0.58), (-1, 0.27), (-1,0.13), (-1,0.066), (-1,0.033), (-1,-0.033)),    
           color='cyan', pointsize=20, zorder=3)
# Add marker for the right angle.
p += disk((-1,0), 0.05, (0,pi/2), color='black', alpha=0.5, zorder=0)
#p += point(((-0.975,0.0215), (0,0)), color='black', pointsize=15, zorder=3)
# Add angle.
p += arc((0.0, 0.0), 0.30, r2=None, sector=(pi,5*pi/6), thickness=1, linestyle='solid', color='black', zorder=4) 
p += text('$30^{^o}$', (-0.225,0.05), fontsize=14, color='black', zorder=4)
# Draw the graph.    
p.plot()
# Save the grap as SVG file.
#p.save('archimedes_figure2.svg')
```
