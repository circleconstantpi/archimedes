```python
# Set max and min values of the co-ordinate axes.
xmin, ymin = -1.1, -1.1
xmax, ymax = 6.3, 1.1 
# Draw an unit circle.
unit_circle = circle((0,0), 1.0, color='blue', xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax,
              axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, fill=False, alpha=1)
unit_circle += circle((0,0), 1.0, color='blue', \
              axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, fill=True, alpha=0.2)
# Draw an right-angled triangle.
triangle = polygon([(0,1.0), (2.0*pi,0), (0,0)], color='red', fill=True, alpha=0.2,                   
              axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, aspect_ratio=1)
triangle += polygon([(0,1), (2*pi,0), (0.0,0.0)], color='red', fill=False, alpha=1, \
              axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, aspect_ratio=1)
# Assemble the graphic.
p = unit_circle + triangle
# Add some text.
p += text(r'$Basis \: des \: rechtwinkligen \: Dreiecks \: ≈ \: 2π$', (pi, -0.5), fontsize=12, color='black', zorder=12)
# Plot the graphic.
p.plot()
```

```python
# Set max and min values of the co-ordinate axes.
xmin, ymin = -1.1, -0.1
xmax, ymax = 6.3, 2.1 

unit_circle = circle((0,1), 1.0, color='blue', xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax, 
                     axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, fill=False, alpha=1)
unit_circle += circle((0,1), 1.0, color='blue', 
                      axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, fill=True, alpha=0.15)
# Draw an right-angled triangle.
triangle = polygon([(0,1.0), (2.0*pi,0), (0,0)], color='red', fill=True, alpha=0.15, 
                   axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, aspect_ratio=1)
triangle += polygon([(0,1), (2*pi,0), (0.0,0.0)], color='red', fill=False, alpha=1, 
                    axes_labels=['$x$','$y$'], thickness=1.5, zorder=0, aspect_ratio=1)
# Assemble the graphic.
p = unit_circle + triangle
# Add some text.
p += text(r'$Basis \: des \: rechtwinkligen \: Dreiecks \: ≈ \: 2π$', (pi, -0.5), fontsize=12, color='black', zorder=12)
p += text(r'$h_{_R}=1$', (0.36, 0.5), fontsize=12, color='black', zorder=20)
p += text(r'$r_{_E}=1$', (0.36, 1.5), fontsize=12, color='black', zorder=20)
p += text(r'$E$', (-0.7, 1.0), fontsize=12, color='black', zorder=20)
p += text(r'$R$', (2.0, 0.35), fontsize=12, color='black', zorder=20)
# Plot the graphic.
p.plot()
```
