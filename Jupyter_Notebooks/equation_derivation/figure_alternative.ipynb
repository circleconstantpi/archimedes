```python
# Radian to degree
def rad2deg(rad):
    '''Convert radian to degree'''
    deg = (180 / pi) * rad
    return(deg)

# Degree to radian
def deg2rad(deg):
    '''Convert degree to radian'''
    rad = (pi / 180) * deg
    return(rad)
```

```python
# Determine the intersection point of an equation of a line with a vertical straight line. 
var('x0')
alpha = 30
for i in range(0,5):
    n = 6 * 2^i
    rad = deg2rad(alpha)
    y0 = -1
    f = -x0*tan(rad) == y0 ;f
    e = f.solve(x0, solution_dict=True)
    x1 = (e[0][x0])/n
    y1 = tan(rad)
    print("alpha =", alpha)
    #print(e)
    #print("x =", x1)
    print("x =", x1.n())
    #print("y =", y1)
    print("y =", y1.n())
    alpha /= 2
```

```python
# Plot a semi-circle with axes.
r = 1
p = plot(sqrt(r^2 - x^2),(x,-1, -0.8665), thickness=1.0, color='magenta', xmin=-1.1, xmax=1.1, 
         ymin=-0.05, ymax=0.6, axes_labels=['$x$','$y$'], zorder=0, aspect_ratio=1)
# Add a straight line.
p += line([(-1.0, 0.0), (-1, 0.58)], color='darkgreen', zorder=0)
# Add the angle bisector.
alpha = 30
for i in range(0,2):
    rad = deg2rad(alpha)
    p += plot(-x*tan(rad),(x, -1.0, 0.0), color='blue', zorder=1)
    alpha /= 2
for i in range(0,3):
    rad = deg2rad(alpha)
    p += plot(-x*tan(rad),(x, -1.0, 0.0), color='purple', zorder=0, linestyle="dashed")
    alpha /= 2
# Add straight line.
p += line([(-0.867, 0.5), (-0.867, 0.0)], color='darkorange', zorder=0)
p += line([(-0.965, 0.260), (-0.965, 0.0)], color='darkorange', zorder=0)
p += line([(-0.965, 0.233), (-0.8665, 0.233)], color='teal', zorder=6)
p += line([(-0.965, 0.233), (-0.965, 0.259)], color='teal', zorder=6)
# Add atriangle.
p += polygon([(-0.965, 0.233),(-0.965, 0.259),(-0.8665, 0.233)], color='springgreen', alpha=0.5, zorder=0)
# Add labeling.  
label = {'$O$': (0.05, -0.05), '$A$': (-1.0, -0.073), '$E$': (-0.83, 0.52), '$F$': (-0.84, 0.26),  '$B$': (-0.8665, -0.073),  
         '$C$': (-1.04, 0.58), '$D$': (-1.04, 0.27), '$G$': (-0.93, 0.28), '$H$': (-0.94, 0.20)} 
for char, posi in label.items():
    p += text(char, posi, fontsize=14, color='black', zorder=2)
# Add points.
p += point(((0,0), (-1,0), (-1, 0.58), (-1, 0.267), (-0.8665, 0.5), (-0.8665, 0.0), (-0.8665, 0.233), (-0.965, 0.259), (-0.965, 0.233)),    
           color='cyan', pointsize=20, zorder=8)
# Add marker for the right angle.
p += disk((-1,0), 0.04, (0,pi/2), color='black', alpha=0.5, zorder=0)
p += disk((-0.8665, 0), 0.04, (0,pi/2), color='black', alpha=0.5, zorder=0)
# Add angle.
p += arc((0.0, 0.0), 0.30, r2=None, sector=(pi,5*pi/6), thickness=1, linestyle='solid', color='black', zorder=4) 
p += text('$30^{^o}$', (-0.225,0.05), fontsize=14, color='black', zorder=4)
# Draw the graph.    
p.plot()
```
