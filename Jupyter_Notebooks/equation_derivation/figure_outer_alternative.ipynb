```python
# ###################################
# Archimedes, Measurement of a Circle
# ###################################
#
# Geometrical construction of an alternative figure
# based on Archimedes outer Figure 3
# Version 0.2
#
# So far, the construction only works for the unit circle.
#
# SageMath 10.1 is using Python 3.8.10.
#
# References
# [1] T. L. HEATH, Sc.D., The Works of Archimedes, EDITED IN MODERN NOTATION, 
# WITH INTRODUCTORY CHAPTERS, CAMBRIDGE: AT THE UNIVERSITY PRESS, 1897
# [Sage] SageMath, the Sage Mathematics Software System (Version 10.1), 
# The Sage Developers, 2019, https://www.sagemath.org.
```

```python
html("""<font color='blue'><b><H1>Archimedes, Measurement of a Circle<br>Construction of a new Figure</H1></b></font><br/>""")
```

```python
# Radian to degree
def rad2deg(rad):
    '''Convert radian to degree'''
    deg = (180 / pi) * rad
    return(deg)

# Degree to radian
def deg2rad(deg):
    '''Convert degree to radian'''
    rad = (pi / 180) * deg
    return(rad)
```

```python
# Plot a partial circle with axes.
r = 1
p = plot(sqrt(r^2 - x^2),(x,-1, -0.8665), thickness=1.0, color='magenta', xmin=-1.1, xmax=1.1, 
         ymin=-0.05, ymax=0.6, axes_labels=['$x$','$y$'], zorder=0, aspect_ratio=1)
# Plot a partial circle without axes.
r = 1/2*sqrt(3)*r
p += plot(sqrt(r^2 - x^2),(x,-0.8665, -0.75), thickness=1.0, color='tomato', zorder=0, aspect_ratio=1)
# Add a straight line.
p += line([(-1.0, 0.0), (-1, 0.58)], color='darkgreen', zorder=0)
# Add the solid angle bisector.
alpha = 30
for i in range(0,2):
    rad = deg2rad(alpha)
    p += plot(-x*tan(rad),(x, -1.0, 0.0), color='blue', zorder=1)
    alpha /= 2
# Add the dashed angle bisector.
for i in range(0,2):
    rad = deg2rad(alpha)
    p += plot(-x*tan(rad),(x, -1.0, 0.0), color='purple', zorder=0, linestyle="dashed")
    alpha /= 2
# Define the co-ordinates of the marked points.
IX = GX = HX = 1/2*sqrt(3) + 1/2*sqrt((sqrt(3)*sqrt(1/(sqrt(3) + 2)^2 + 1) - 2)^2 - 
               (sqrt(3)/(sqrt(3) + 2) - 2/((sqrt(3) + 2)*sqrt(1/(sqrt(3) + 2)^2 + 1)))^2)
O = (0, 0)
A = (-1, 0)
B = (-1/2*sqrt(3), 0)
C = (-1, 1/sqrt(3))
D = (-1, sqrt(3)/(2*sqrt(3) + 3))
E = (-1/2*sqrt(3), 1/2)
F = (-1/2*sqrt(3), 1/2*sqrt(3)/(sqrt(3) + 2)) 
G = (-GX, 1/2/((sqrt(3) + 2)*sqrt((sqrt(3) + 2)/(4*sqrt(3) + 7))))
H = (-HX, 1/2*sqrt(3)/(sqrt(3) + 2)) 
I = (-IX, 0)
# Add some straight line.
p += line([(F[0], F[1]), (-HX, H[1])], color='teal', zorder=6)
p += line([(-GX, G[1]), (-HX, H[1])], color='teal', zorder=6)
p += line([(-GX, G[1]), (-IX, I[1])], color='darkorange', zorder=0)
p += line([(B[0], B[1]), (E[0], E[1])], color='darkorange', zorder=0) 
p += line([(-0.989, 0.130), (-0.989, 0.0)], color='darkorange', zorder=0)
# Add triangles.
p += polygon([(-0.965, 0.233), (-0.965, 0.259), (-0.866, 0.233)], color='springgreen', alpha=0.5, zorder=0, fill= True)
p += polygon([(-0.989, 0.114), (-0.989, 0.130), (-0.866, 0.114)], color='springgreen', alpha=0.5, zorder=0, fill= True)
p += polygon([(-0.989, 0.114), (-0.989, 0.130), (-0.866, 0.114)], color='teal', alpha=1, zorder=0, fill= False)
p += polygon([(-0.997, 0.056), (-0.997, 0.065), (-0.866, 0.056)], color='springgreen', alpha=0.5, zorder=0, fill= True)
p += polygon([(-0.997, 0.056), (-0.997, 0.065), (-0.866, 0.056)], color='teal', alpha=1, zorder=0, fill= False)
# Add the labeling.  
label = {'$O$': (0.05, -0.05), '$A$': (-1.0, -0.073), '$E$': (-0.83, 0.52), r'$F\,=f$': (-0.809, 0.26),  '$B$': (-0.8665, -0.073),  
         '$C$': (-1.04, 0.58), '$D$': (-1.04, 0.27), '$g$': (-0.943, 0.28), '$h$': (-0.946, 0.205), '$i$': (-0.954, -0.03)} 
for char, posi in label.items():
    p += text(char, posi, fontsize=14, color='black', zorder=2)
# Add colored points.
p += point((O, A, C, E, B, F, D, I, G, H), color='cyan', pointsize=20, zorder=8)
# Add marker for the right angle.
p += disk((-1,0), 0.04, (0,pi/2), color='black', alpha=0.5, zorder=0)
p += disk((-0.8665, 0), 0.04, (0,pi/2), color='black', alpha=0.5, zorder=0)
# Add the 30 angle.
p += arc((0.0, 0.0), 0.30, r2=None, sector=(pi, 5*pi/6), thickness=1, linestyle='solid', color='black', zorder=4) 
p += text('$30^{^o}$', (-0.225,0.050), fontsize=14, color='black', zorder=10)
# Draw the graph.    
p.plot()
```
