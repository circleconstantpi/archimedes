```python
COL = sorted(colors)
count = 0
num = int((len(COL)/6).n())+1
for i in range(0,num):
    a = polygon([(0,0), (0,1), (1,1), (1,0)], color=COL[count], axes=False)
    a += text(COL[count], (0.5,0.5), fontsize=24, color='black')
    for j in range(1,6):
        try:
            a += polygon([(0+j,0), (0+j,1), (1+j,1), (1+j,0)], color=COL[count+j], axes=False)
            a += text(COL[count+j], (0.5+j,0.5),fontsize=24, color='black')
        except:
            a += polygon([(0+j,0), (0+j,1), (1+j,1), (1+j,0)], color='white', axes=False)
            a += text('intentionally left blank', (0.5+j,0.5),fontsize=24, color='black')
    a.show(figsize=32, aspect_ratio=1)
    count+=6  
```
